## Question 1: Strings, Lists & List Comprehensions

**Explicit Tasks**:
1. Strip whitespace and convert the entire string to lowercase.
2. Remove the characters `!` and `.`.
3. Split the text into a list of words.
4. Use a list comprehension to keep only words with **4 or more letters**.
5. Use another list comprehension to reverse each word from step 4.
6. Convert the reversed list into a set to remove duplicates.
7. Print all steps and the final count of unique reversed words.

````python
# SOLUTION - Question 1
text = "  The quick brown fox jumps over the lazy dog! Python is awesome for data science in 2024.  "

# 1. Clean
cleaned = text.strip().lower()

# 2. Remove punctuation
cleaned = cleaned.replace('!', '').replace('.', '')

# 3. Split
words = cleaned.split()

# 4. Filter long words
long_words = [w for w in words if len(w) >= 4]

# 5. Reverse
reversed_words = [w[::-1] for w in long_words]

# 6. Unique set
unique_reversed = set(reversed_words)

print("Words:", words)
print("Long words:", long_words)
print("Reversed:", reversed_words)
print("Unique reversed:", unique_reversed)
print("Final count:", len(unique_reversed))
````

## Question 2: Dictionaries (Aggregation & Grouping)

**Explicit Tasks**:
1. Build a dictionary with each person’s **average score** (rounded to 1 decimal).
2. Identify and print the person with the highest average.
3. Create a dictionary grouping people by their average score (score → list of names).
4. Create a set containing all unique average scores.

````python
# SOLUTION - Question 2
from collections import defaultdict

scores = [('Alice', 85), ('Bob', 92), ('Charlie', 78), ('Alice', 95), ('Bob', 88)]

totals = defaultdict(int)
counts = defaultdict(int)

for name, score in scores:
    totals[name] += score
    counts[name] += 1

avg_scores = {name: round(totals[name] / counts[name], 1) for name in totals}

print("Average Scores:", avg_scores)
print("Highest scorer:", max(avg_scores, key=avg_scores.get))

grouped = defaultdict(list)
for name, avg in avg_scores.items():
    grouped[avg].append(name)
print("Grouped by score:", dict(grouped))
print("Unique averages:", set(avg_scores.values()))
````

## Question 3: Sets

**Explicit Tasks**:

2. Compute and print: intersection, union, difference (list1 - list2), symmetric difference.
3. Print the length of each resulting set.

````python
# SOLUTION - Question 3
list1 = ["python", "data", "ml", "ai", "python", "sql", "data"]
list2 = ["data", "sql", "tableau", "powerbi", "ai", "ml", "python"]

s1 = set(list1)
s2 = set(list2)

print("Intersection:", s1 & s2, "|", "Size:", len(s1 & s2))
print("Union:", s1 | s2, "|", "Size:", len(s1 | s2))
print("Only in list1:", s1 - s2, "|", "Size:", len(s1 - s2))
print("Symmetric difference:", s1 ^ s2, "|", "Size:", len(s1 ^ s2))
````

## Question 4: Functions + Conditionals + Loops

**Explicit Tasks**:
Write a function `calculate_stats(numbers)` that returns a dictionary containing exactly these 7 keys:
- positive_count, negative_count, zero_count
- even_sum, odd_sum
- max_value, min_value

Use **only loops and if/elif/else**. Do not use `sum()`, `max()`, or `min()`.

````python
# SOLUTION - Question 4
def calculate_stats(numbers):
    positive = negative = zero = 0
    even_sum = odd_sum = 0
    if not numbers:
        return {}
    max_val = min_val = numbers[0]
    
    for n in numbers:
        if n > 0:
            positive += 1
        elif n < 0:
            negative += 1
        else:
            zero += 1
        
        if n % 2 == 0:
            even_sum += n
        else:
            odd_sum += n
        
        if n > max_val:
            max_val = n
        if n < min_val:
            min_val = n
    
    return {
        'positive_count': positive,
        'negative_count': negative,
        'zero_count': zero,
        'even_sum': even_sum,
        'odd_sum': odd_sum,
        'max_value': max_val,
        'min_value': min_val
    }

nums = [5, -3, 0, 8, -1, 4, 0, 12, -7, 3]
print(calculate_stats(nums))
````

## Question 5: Debugging & Syntax

**Explicit Tasks**:
Fix all syntax and logical bugs in the given code and extend it to also track the **count** of each fruit.

````python
# SOLUTION - Question 5
data = ["apple:2", "banana:3", "apple:1", "cherry:5", "BANANA:2"]
result = {}
count = {}

for item in data:
    try:
        fruit, qty_str = item.split(":")
        qty = int(qty_str)
        fruit_lower = fruit.lower()
        
        result[fruit_lower] = result.get(fruit_lower, 0) + qty
        count[fruit_lower] = count.get(fruit_lower, 0) + 1
    except Exception as e:
        print(f"Error processing '{item}': {e}")

print("Total quantities:", result)
print("Occurrence counts:", count)
````

## Question 7: Pandas Data Cleaning

**Explicit Tasks**:
1. Strip whitespace and Title Case the name column.
2. Convert age to numeric (coerce errors).
3. Fill missing cities with “Unknown” and Title Case.
4. Drop duplicate names.
5. Show final DataFrame and `.info()`.

````python
# SOLUTION - Question 7
import pandas as pd

dirty_data = pd.DataFrame({
    "name": [" Alice", "bob ", "Charlie", "  alice", "Eve"],
    "age": [25, "30", 28, None, 35],
    "city": ["New York", "new york", "Los Angeles", "Chicago", None]
})

clean = dirty_data.copy()
clean['name'] = clean['name'].str.strip().str.title()
clean['age'] = pd.to_numeric(clean['age'], errors='coerce')
clean['city'] = clean['city'].fillna("Unknown").str.title()
clean = clean.drop_duplicates(subset=['name'])

print(clean)
print("\nInfo:")
print(clean.info())
````

## Question 8: GroupBy & Aggregation

**Explicit Tasks**:
1. Group by region and category.
2. Aggregate: total_sales, avg_quantity, count, total_quantity.
3. Add sales_per_unit column.
4. Reset index and display.

````python
# SOLUTION - Question 8
import pandas as pd

sales = pd.DataFrame({
    "region": ["North", "South", "North", "East", "South", "East", "North"],
    "category": ["A", "A", "B", "B", "A", "B", "A"],
    "sales": [100, 150, 200, 120, 180, 90, 220],
    "quantity": [10, 15, 8, 12, 20, 5, 25]
})

grouped = (sales.groupby(['region', 'category'])
           .agg(total_sales=('sales', 'sum'),
                avg_quantity=('quantity', 'mean'),
                transaction_count=('sales', 'count'),
                total_quantity=('quantity', 'sum'))
           .reset_index())

grouped['sales_per_unit'] = grouped['total_sales'] / grouped['total_quantity']
print(grouped)
````

## Question 9: Pandas Merge + Feature Engineering

**Explicit Tasks**:
1. Left merge customers and orders on customer_id.
2. Create boolean column `high_value` (amount > 200).
3. Group by city: total_amount and high_value_count.
4. Print merged table and summary.

````python
# SOLUTION - Question 9
import pandas as pd

customers = pd.DataFrame({"customer_id": [1,2,3,4], "name": ["Alice","Bob","Charlie","Dana"], "city": ["NY","LA","NY","Chicago"]})
orders = pd.DataFrame({"customer_id": [1,2,1,3,5], "amount": [100,200,150,300,400], "date": ["2024-01","2024-02","2024-01","2024-03","2024-01"]})

merged = pd.merge(customers, orders, on='customer_id', how='left')
merged['high_value'] = merged['amount'] > 200

summary = merged.groupby('city').agg(
    total_amount=('amount', 'sum'),
    high_value_count=('high_value', 'sum')
).reset_index()

print("Merged Data:\n", merged)
print("\nCity Summary:\n", summary)
````

## Question 10: The Ultimate Challenge (Very Hard)

**Explicit Tasks**:
1. Write `parse_log(log)` function with robust parsing and error handling.
2. Convert timestamp to datetime and amount to float when possible.
3. Build clean DataFrame.
4. Add `is_purchase` column.
5. Return summary: unique users, total revenue, top user, most common action, error count.

````python
# SOLUTION - Question 10
import pandas as pd

def parse_log(log):
    record = {}
    try:
        parts = log.split()
        for part in parts:
            if '=' in part:
                key, value = part.split('=', 1)
                record[key] = value
        
        # Type conversion
        if 'timestamp' in record:
            try:
                record['timestamp'] = pd.to_datetime(record['timestamp'])
            except:
                record['timestamp'] = None
        if 'amount' in record:
            try:
                record['amount'] = float(record['amount'])
            except:
                record['amount'] = None
        
        record['is_purchase'] = record.get('action') == 'purchase'
        return record
    except:
        return None

logs = [
    "user=alice action=login timestamp=2024-01-01 10:00:00 ip=192.168.1.1",
    "user=bob action=purchase amount=99.99 item=laptop timestamp=2024-01-02",
    "user=alice action=logout timestamp=2024-01-01 11:00:00",
    "user=charlie action=login timestamp=bad-date ip=10.0.0.1",
    "user=bob action=purchase amount=abc item=mouse",
    "user=alice action=login timestamp=2024-01-03 09:00:00 ip=192.168.1.1",
    "user=dana action=view timestamp=2024-01-04 14:00:00 ip=172.16.0.1"
]

parsed = [parse_log(log) for log in logs]
df = pd.DataFrame([p for p in parsed if p is not None])

summary = {
    "total_unique_users": df['user'].nunique(),
    "total_revenue": df['amount'].sum(),
    "top_user_by_revenue": df.groupby('user')['amount'].sum().idxmax() if not df.empty else None,
    "most_common_action": df['action'].mode()[0] if not df.empty else None,
    "error_count": len(logs) - len(df)
}

print("Final Summary:", summary)
df
````
